## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# your code here
def normal_dist(mu, sigma, count):
    return norm.rvs(loc=mu, scale=sigma, size=count)
```

```python
# your code here
normal1 = normal_dist(10, 1, 1000)
```

```python
# your code here
normal2 = normal_dist(10, 50, 1000)
```

```python
# your code here
# graph two  normal distributions side by side, sharing the same y axis but each with their own x axis
bins = 100
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].hist(normal1, bins=bins)
ax[1].hist(normal2, bins=bins)
plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# The standard deviation changes, so ones values are more spread out than the other.
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
def exponential (count):
    return np.random.exponential(1/10, count)
```

```python
exp1 = exponential(10)
exp2 = exponential(100)
```

```python
# graph exp1 and exp2 with histrograms and 100 bins, and sharing the same y axis
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].hist(exp1, bins=100)
ax[1].hist(exp2, bins=100)
plt.tight_layout()
plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
def exp_cdf (mean, x):
    return 1 - np.exp(-x/mean)
```

```python
# your answer here
# Hint: This is same as saying P(x<15)
print(f"The probability of a customer spending less than 15 minutes in a bank is {exp_cdf(10, 15)}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here
print(f"The probability of a customer spending more than 15 minutes in a bank is {1- exp_cdf(10, 15)}")
```
